```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.epochtimes.com/b5/20/10/28/n12509290.htm")
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
sel = soup.select("div.post_content p")
```

```python
for s in sel:
    if "玫瑰花" in s.text:
        print(s.text)
```
